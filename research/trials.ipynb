```python
from langchain.document_loaders import PyPDFLoader,DirectoryLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
```

```python
pdf_path="data/3-Lecture-Common-Skin-Disease-2019.pdf"  # Removed leading slash for Windows compatibility
def load_and_split_data(pdf_path):
    loader = PyPDFLoader(pdf_path)
    documents = loader.load()  # Fixed method name to lowercase
    text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
    texts = text_splitter.split_documents(documents)  # Split the documents
    return texts  # Return the split texts

extracted_data=load_and_split_data(pdf_path)
extracted_data # Display the first chunk of extracted data
```

```python
print(len(extracted_data))
```

```python
from typing import List 
from langchain.schema import    Document

def filter_to_minimal_chunk(extrac)->List[Document]:
    minimal_docs:List[Document]=[]
    for doc in extracted_data:
        minimal_docs.append(
            Document(
                page_content=doc.page_content,
                metadata={"source":doc.metadata.get("source")}
                )
            
            )
    return minimal_docs
```

```python
type(List[Document])
```

```python
filtered=filter_to_minimal_chunk(extracted_data)
filtered
```

```python
#split chunk
def split(minimal_doc):
    text_slitter=RecursiveCharacterTextSplitter(
        chunk_size=500,
        chunk_overlap=20
        
    )
    text_chunk= text_slitter.split_documents(minimal_doc)
    return text_chunk
```

```python
text_chunk=split(filtered)
```

```python
from langchain_openai import ChatOpenAI
chat_model=ChatOpenAI(
    model="gpt-4o"
)
```

```python
from langchain.chains import create_retrieval_chain
from langchain.chains.combine_documents import create_stuff_documents_chain
from langchain_core.prompts import ChatPromptTemplate
```

```python
system_prompt=(
   "you are a medical chatbot."
   "use the following retrieved context to answer."
   "if you do not know the answer say thta you do not know"
"if you know the answer provide three sentences with maximum similarities"
"\n\n"
"{context}"
)


prompt=ChatPromptTemplate.from_messages(
    [
        ("system",system_prompt),
        ("human","{input}")
    ]
)
```

```python
question_answer_chain=create_stuff_documents_chain(chat_model,prompt)
rag_chain=create_retrieval_chain(retriever,question_answer_chain)
```

```python
response=rag_chain.invoke({"input":"what is rosacea?"})
print(response["answer"])
```

```python
response=rag_chain.invoke({"input":"what are the treatments for  rosacea?"})
print(response["answer"])
```
